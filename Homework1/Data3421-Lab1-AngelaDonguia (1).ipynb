#### 5. Which one of these independent variables is positively correlated with dependent variables?

```python
df[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']].corr()['Profit']
df.plot.scatter(x='R&D Spend',y='Profit')
df.plot.scatter(x='Marketing Spend',y='Profit')
df.plot.scatter(x='Administration',y='Profit')
```

based on this we can determine that R&D Spend is positively correlated with Profit
