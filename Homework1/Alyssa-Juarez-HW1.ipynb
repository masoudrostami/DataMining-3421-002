## 3. Create a histogram and boxplot for the independent variables to understand the distribution and skewness (mention which one has right or left skewness or has normal distribution)?

```python
data_1 = df['R&D Spend']
data_2 = df['Administration']
data_3 = df['Marketing Spend']
data_4 = df['Profit'] 
#went ahead and did profit too since it is a numerical value even though it 
#is the dependent in this circumstance
```

```python
#R&D Spend
plt.hist(data_1, bins = 15)
plt.show()

#Other descriptive stats
print("min:", data_1.min())
print("max:", data_1.max())
print("mean:", np.mean(data_1))
print("median:", np.median(data_1))
print("standard deviation:", np.std(data_1))

#Administration
plt.hist(data_2, bins = 15)
plt.show()

#Other descriptive stats
print("min:", data_2.min())
print("max:", data_2.max())
print("mean:", np.mean(data_2))
print("median:", np.median(data_2))
print("standard deviation:", np.std(data_2))

#Marketing Spend
plt.hist(data_3, bins = 15)
plt.show()

#Other descriptive stats
print("min:", data_3.min())
print("max:", data_3.max())
print("mean:", np.mean(data_3))
print("median:", np.median(data_3))
print("standard deviation:", np.std(data_3))

#Profit
plt.hist(data_4, bins = 15)
plt.show()

#Other descriptive stats
print("min:", data_4.min())
print("max:", data_4.max())
print("mean:", np.mean(data_4))
print("median:", np.median(data_4))
print("standard deviation:", np.std(data_4))
```

```python
#Data points
data_1 = df['R&D Spend']
data_2 = df['Administration']
data_3 = df['Marketing Spend']
data_4 = df['Profit'] 
data = [data_1, data_2, data_3, data_4]
 
fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)
 
# Creating axes instance
bp = ax.boxplot(data, patch_artist = True,
                notch ='True', vert = 0)
 
for median in bp['medians']:
    median.set(color ='red',
               linewidth = 3)
 # x-axis labels
ax.set_yticklabels(['R&D Spend', 'Administration',
                    'Marketing Spend', 'Profit'])
 
# Adding title
plt.title("Customized box plot")
 
# Removing top axes and right axes
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
     
# show plot
plt.show()
```
