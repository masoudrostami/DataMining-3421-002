#### 3. Create histogram and boxplot for independent variables to understand the distribution and skewness (mention which one has right or left skewness or has normal distribution)?

```python
fig = plt.figure(figsize = (19,9))
# histogram plot items sold for each item 
sns.histplot(data = df,x = "R&D Spend")
# GRAPH Title, x-axis label, y-axis label 
plt.title("R&D Spend",fontsize=15)
plt.xlabel("R&D Spend",fontsize=15)
plt.show()
```

```python
fig = plt.figure(figsize = (19,9))
# histogram plot items sold for each item 
sns.histplot(data = df,x = "Marketing Spend")
# GRAPH Title, x-axis label, y-axis label 
plt.title("Marketing Spend",fontsize=15)
plt.xlabel("Marketing Spend",fontsize=15)
plt.show()
```

```python
fig = plt.figure(figsize = (19,9))
# histogram plot items sold for each item 
sns.histplot(data = df,x = "Administration")
# GRAPH Title, x-axis label, y-axis label 
plt.title("Administration",fontsize=15)
plt.xlabel("Administration",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,4))
df.boxplot(column=['R&D Spend','Administration','Marketing Spend'])
```
