#### 3. Create histogram and boxplot for independent variables to understand the distribution and skewness (mention which one has right or left skewness or has normal distribution)?

```python
# import libraries and packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
rd=df["R&D Spend"]
rd.hist(bins=20)
plt.axvline(rd.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(rd.median(), color='r', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(rd.mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(rd.mean()))

plt.text(rd.mean()*1.1, max_ylim*0.85, 'Median: {:.2f}'.format(rd.median()),color='r')
if rd.mean()== rd.median():
    plt.text(rd.mean()*1.1, max_ylim*0.80, 'Normal Distribution',color='g')
if rd.mean()>= rd.median():
    plt.text(rd.mean()*1.1, max_ylim*0.80, 'Right Skewness',color='g')
if rd.mean()<= rd.median():
    plt.text(rd.mean()*1.1, max_ylim*0.80, 'Left Skewness',color='g')
plt.text(rd.mean()*1.1, max_ylim*0.75, 'Std Dev: {:.2f}'.format(rd.std()))    

# Set x-axis label
plt.title('R&D Spend (US Dollars ($))')
plt.show()

adm=df["Administration"]
adm.hist(bins=20)
plt.axvline(adm.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(adm.median(), color='r', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(adm.mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(adm.mean()))

plt.text(adm.mean()*1.1, max_ylim*0.85, 'Median: {:.2f}'.format(adm.median()),color='r')
if adm.mean()== adm.median():
    plt.text(adm.mean()*1.1, max_ylim*0.80, 'Normal Distribution',color='g')
if adm.mean()>= adm.median():
    plt.text(adm.mean()*1.1, max_ylim*0.80, 'Right Skewness',color='g')
if adm.mean()<= adm.median():
    plt.text(adm.mean()*1.1, max_ylim*0.80, 'Left Skewness',color='g')
plt.text(adm.mean()*1.1, max_ylim*0.75, 'Std Dev: {:.2f}'.format(adm.std()))    

# Set x-axis label
plt.title('Administration Expense (US Dollars ($))')
plt.show()

mkt=df["Marketing Spend"]
mkt.hist(bins=20)
plt.axvline(mkt.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(mkt.median(), color='r', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(mkt.mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(mkt.mean()))

plt.text(mkt.mean()*1.1, max_ylim*0.85, 'Median: {:.2f}'.format(mkt.median()),color='r')
if mkt.mean()== mkt.median():
    plt.text(mkt.mean()*1.1, max_ylim*0.80, 'Normal Distribution',color='g')
if mkt.mean()>= mkt.median():
    plt.text(mkt.mean()*1.1, max_ylim*0.80, 'Right Skewness',color='g')
if mkt.mean()<= mkt.median():
    plt.text(mkt.mean()*1.1, max_ylim*0.80, 'Left Skewness',color='g')
plt.text(mkt.mean()*1.1, max_ylim*0.75, 'Std Dev: {:.2f}'.format(mkt.std()))    

# Set x-axis label
plt.title('Marketing Spend (US Dollars ($))')
plt.show()

state=df["State"]
state.hist(bins=3,rwidth=0.8,align='mid')

plt.show()

#df.boxplot()
sns.boxplot(data=df)
```
