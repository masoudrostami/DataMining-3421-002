```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as st
from statistics import mode
from tabulate import tabulate
import seaborn as sns

with open("Startups.csv", 'r') as file:
    startups = pd.read_csv(file)
```

#### 1. Show datatype for each variable and summary of the variables?

```python
types = list()
col_names = list(startups.columns.values)
col_descriptions = ["Research and Development Expenses","Administration Expenses", "Marketing Expenses", "State Locality", "Revenue - Cost = "]

for i in startups.loc[0]:
    types.append(str(type(i)))
    
table = list(zip(col_names, col_descriptions, types))
print(tabulate(table, tablefmt="heavy_outline", headers=['Variable', 'Description', 'Type']))
```

#### 2. Show the descriptive statistics for the data frame?

```python
qt = list()
ql = list()

floatType = type(startups.loc[0][0])
for colName, colData in startups.iteritems():
    if type(colData[0]) == floatType:
        qt.append(colName)
        qt.append(np.mean(colData)),
        qt.append(np.median(colData)),
        qt.append(min(colData)),
        qt.append(max(colData) - min(colData)),
        qt.append(max(colData)),
        qt.append(np.var(colData)),
        qt.append(np.std(colData)),
        qt.append(np.percentile(colData, 25, interpolation = 'midpoint')),
        qt.append(st.iqr(colData, interpolation='midpoint')),
        qt.append(np.percentile(colData, 75, interpolation = 'midpoint'))
    else:
        ql.append(colName)
        ql.append(colData.value_counts() / len(colData)),
        ql.append(mode(colData))

qt = list(zip(*(iter(qt),) * 11))
ql = list(zip(*(iter(ql),) * 1))

print(f"Qualitative Variable Breakdown is {ql}.")
print(tabulate(qt, tablefmt="heavy_outline", headers=['Quantitative', 'mean', 'median', 'minimum', 'ranges', 'maximum', 'variance', 'stdev', 'q1', 'iqr', 'q3']))        
#print(tabulate(ql, tablefmt="heavy_outline", headers=['Qualitative', 'Percentages', 'Mode']))
                     
        
```

#### 3. Create histogram and boxplot for independent variables to understand the distribution and skewness (mention which one has right or left skewness or has normal distribution)?

```python
startups_indep = startups.loc[:,startups.columns != 'Profit']
def plotting(dataset):
    index = 1
    for column_name in dataset:
        plt.figure(figsize=(15,20))
        plt.subplot(4,2,index)
        plt.hist(startups_indep[column_name])
        plt.title(column_name.upper())
        index += 1
        if index < 7:
            plt.subplot(4,2,index)
            plt.boxplot(startups_indep[column_name])
            plt.title(column_name.upper())
            index += 1
        plt.show()
        
plotting(startups_indep)
```

All boxplots show a normal distribution.
