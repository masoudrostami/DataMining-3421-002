#### 3. Create histogram and boxplot for independent variables to understand the distribution and skewness (mention which one has right or left skewness or has normal distribution)?

```python
##'R&D Spend', 'Administration', 'Marketing Spend'
def plot_histogram(df, column):
    plt.figure(figsize = (8, 8))
    plt.hist(df[column], bins = 12, alpha = 1, edgecolor = 'k')
    plt.title(column + " Distribution")
    plt.xlabel("$")
    plt.ylabel("Frequency")
    plt.show()
```

```python
plot_histogram(df, 'R&D Spend')
```

```python
plot_histogram(df, 'Administration')
```

```python
plot_histogram(df, 'Marketing Spend')
```

```python
independent_dict = dict()
independent_dict['R&D Spend'] = df['R&D Spend']
independent_dict['Administration'] = df['Administration']
independent_dict['Marketing Spend'] = df['Marketing Spend']

fig, ax = plt.subplots(figsize = (8, 8))
ax.boxplot(independent_dict.values(), showfliers = True, vert = True, showmeans = True)
ax.set_xticklabels(independent_dict.keys())
ax.set_ylabel("Price ($K)")
plt.title("Distribution of Price")
plt.show()
```

```python
#Skewness:
#If the skewness is between -0.5 & 0.5, the data are nearly symmetrical.
#If the skewness is between -1 & -0.5 (negative/left skewed) or between 0.5 & 1(positive/right skewed), the data are slightly skewed.
#If the skewness is lower than -1 (negative skewed) or greater than 1 (positive skewed), the data are extremely skewed.

#Kurtosis:
#Mesokurtic distribution: normal distribution tails (=3)
#Platykurtic distribution: tail is thinner (Data is more centered in the middle than Normal Distribution) (<3)
#Leptokurtic distribution: tail is fatter (Data is more spread out than Normal Distribution) (>3)

#Shapiro Wilk test:
#If p-value of Shapiro Wilk Test > 0.05, we assume a normal distribution
#If p-value of Shapiro Wilk Test < 0.05, we do not assume a normal distribution

skew1 = np.round(stats.skew(df['R&D Spend']), 2)
skew2 = np.round(stats.skew(df['Administration']), 2)
skew3 = np.round(stats.skew(df['Marketing Spend']), 2)
skew_array = np.array([[skew1, skew2, skew3]])
    
kurtosis1 = np.round(stats.kurtosis(df['R&D Spend'], fisher = False), 2)
kurtosis2 = np.round(stats.kurtosis(df['Administration'], fisher = False), 2)
kurtosis3 = np.round(stats.kurtosis(df['Marketing Spend'], fisher = False), 2)
kurtosis_array = np.array([[kurtosis1, kurtosis2, kurtosis3]])

shapiro1 = stats.shapiro(df['R&D Spend'])[1]
shapiro2 = stats.shapiro(df['Administration'])[1]
shapiro3 = stats.shapiro(df['Marketing Spend'])[1]
shapiro_array = np.array([[shapiro1, shapiro2, shapiro3]])



statistics = np.concatenate((skew_array, kurtosis_array, shapiro_array), axis = 0)
statistics
```

```python
results_df = pd.DataFrame(statistics, index = ['Skewness', 'Kurtosis', 'Shapiro'], columns = ['R&D Spend', 'Administration', 'Marketing Spend'])
results_df
```

```python
#R&D spending: The distribution is assumed Normal and symmetrical, but the tails are a bit thinner than a Normal Distribution.
#Administration: Even though the distribution is assumed Normal based on Shapiro Wilk test, it is slightly negatively skewed.
#Still the tails are same as a Normal distribution.
#Marketing Spend: The distribution is assumed Normal and symmetrical, but the tails are a bit thinner than a Normal Distribution.
```

#### 6. Which one of the States has highest profit?

```python
all_states = df['State'].unique()
profit_dict = dict()
for state in all_states:
    total = df.loc[df['State'] == state]['Profit'].sum()
    profit_dict[state] = np.round(total, 2)
print(profit_dict)
print('The state with maximum profit: ', max(profit_dict, key = profit_dict.get))
```

#### 7. Will a company that spends more on marketing perform better (higher profit) or a company spends less on marketing?

```python
##Visualization
m, b = np.polyfit(df['Marketing Spend'], df['Profit'], 1)
plt.figure(figsize = (8, 8))
plt.scatter(df['Marketing Spend'], df['Profit'])
plt.plot(df['Marketing Spend'], m*df['Marketing Spend'] + b, color = 'orange')
plt.title('Marketing Spend vs. Profit')
plt.xlabel("Marketing Spend ($)")
plt.ylabel("Profit ($)")
plt.show()

##Based on the graph and correlation = 0.747766, we can say that there is a strong positive correlation between
##Marketing Spend and Profit. The more a company spends on Marketing, the more profit they made.
```

```python
##This represents the company that spend most of the budget on marketing
df_marketing_most = df.loc[(df['Marketing Spend'] >= df['R&D Spend']) & (df['Marketing Spend'] >= df['Administration'])]

##This represents the company that has the second highest budget on marketing
df_marketing_partial = df.loc[(df['Marketing Spend'] >= df['R&D Spend']) ^ (df['Marketing Spend'] >= df['Administration'])]

##This represents the company that spends the least on marketing
df_marketing_least = df.loc[(df['Marketing Spend'] < df['R&D Spend']) & (df['Marketing Spend'] < df['Administration'])]
```

```python
mean1 = np.round(df_marketing_most['Profit'].mean(), 2)
mean2 = np.round(df_marketing_partial['Profit'].mean(), 2)
mean3 = np.round(df_marketing_least['Profit'].mean(), 2)
mean_array = np.array([[mean1, mean2, mean3]])
    
median1 = np.round(df_marketing_most['Profit'].median(), 2)
median2 = np.round(df_marketing_partial['Profit'].median(), 2)
median3 = np.round(df_marketing_least['Profit'].median(), 2)
median_array = np.array([[median1, median2, median3]])

statistics = np.concatenate((mean_array, median_array), axis = 0)
results_df = pd.DataFrame(statistics, index = ['Mean', 'Median'], 
                          columns = ['1st Marketing', '2nd Marketing', '3rd Marketing'])
results_df

##Based on the result, if a company that spends most of its budget on marketing, it achieves the most profit. 
##However, if marketing is a 2nd most spending, then it achieves the least profit.
##Companies that spend least on marketing perform better than companies with 2nd most spending on marketing.
```

```python
columns = ['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']
n_columns = 4
fig, ax = plt.subplots(n_columns, n_columns, figsize = (15, 20))
df_NY = df.loc[df['State'] == 'New York']
df_Cali = df.loc[df['State'] == 'California']
df_Fl = df.loc[df['State'] == 'Florida']
for i in range(n_columns):
    for j in range(n_columns):
        if i == j:
            ax[i][j].hist(df_NY[columns[i]], bins = 6, alpha = 0.5,
                          color = 'green', ec = 'black', density = False, label = 'New York')
            ax[i][j].hist(df_Cali[columns[i]], bins = 6, alpha = 0.5, 
                          density = False, color = 'blue', ec = 'black', label = 'California')
            ax[i][j].hist(df_Fl[columns[i]], bins = 6, alpha = 0.5, 
                          density = False, color = 'orange', ec = 'black', label = 'Florida')
            ax[i][j].set_xlabel(columns[i], loc = 'right')
            ax[i][j].set_ylabel("Frequency", loc = 'top')    
            if (i == 0) and (j == 0):
                ax[i][j].legend(loc = "upper right")
        else:
            colors = {'New York': 'green', 'California' : 'blue', 'Florida': 'orange'}
            ax[i][j].scatter(df[columns[i]], df[columns[j]], c = df['State'].map(colors))
            ax[i][j].set_xlabel(columns[i], loc = 'right')
            ax[i][j].set_ylabel(columns[j], loc = 'top') 

fig.tight_layout()
plt.show()
```
