# Create histogram and boxplot for independent variables to understand the distribution and skewness (mention which one has right or left skewness or has normal distribution)?

```python
plt.figure(figsize = (5,5))
plt.hist(sps['R&D Spend'], bins=12, edgecolor = 'k')
plt.title('R&D Spend')
plt.show()
```

```python
plt.figure(figsize = (5,5))
plt.hist(sps['Administration'], bins=12, edgecolor = 'k')
plt.title('Administration')
plt.show()
```

```python
plt.figure(figsize = (5,5))
plt.hist(sps['Marketing Spend'], bins=12, edgecolor = 'k')
plt.title('Marketing Spend')
plt.show()
```

```python
plt.figure(figsize = (5,10))
sps.boxplot(column = ['R&D Spend','Administration','Marketing Spend'])
```

R&D Spend is left-skewed/negatively-skewed as the median of it lies more towards the top end of the box.
Administration has normal distribution as the median lies in between hence it has no skew.
And marketing spend has a positive distribution as it is right-skewed

# Will a company that spends more on marketing perform better (higher profit) or a company spends less on marketing?


```python
sps.plot.scatter('R&D Spend','Profit', figsize = (5,5))
```

```python
sps.plot.scatter('Marketing Spend','Profit', figsize = (5,5))
```

```python
sps.plot.scatter('Administration','Profit', figsize = (5,5))
```

Based on the scatterplot, it seems that the more money spend on R&D the better profit it yields compared to marketing. So, a company that spends more on marketing will do poor than a company that spends less on marketing
