## 2. Show the descriptive statistic for the data frame?

```python
for i in col_names[:4]:
    print(data[i].describe())
```

## 3. Create histogram and boxplot for independent variables to understand the distribution and skewness (mention which one has right or left skewness or has normal distribution)?

```python
for i in col_names[4:]: #This will total the total amount in each of the categorical features
    print(i+' Total: {}'.format(sum(data[i])))
```

```python
for i,feature_name in enumerate(col_names):
    plt.figure(figsize=(20,20))
    plt.subplot(5,3,i+2)
    plt.hist(data[feature_name])
    plt.xlabel(feature_name)
    plt.ylabel('Frequency')
    plt.show()
    
```

R&D spend doesn't seem to be skewed any type of direction very much nor does it look normally distrubuted.
Administration seems to be slightly left skewed or normally distributed.
Marketing spend seems to be right skewed.
Profit seems to be the most normally distributed out of the bunch. 

```python
for i,feature_name in enumerate(col_names[:4]):
    plt.figure(figsize=(20,20))
    plt.subplot(5,3,i+2)
    plt.boxplot(data[feature_name])
    plt.xlabel(feature_name)
    plt.ylabel('Frequency')
    plt.show() 
```

## 4. Which one of independent variables has the highest variability?

```python
for i in col_names[:3]:
    print(i+' Standard Deviation: {}'.format(data[i].std()))
```

As We can see Marketing Spend has the highest variability by far.

## 5. Which one of these independent variables is positively correlated with dependent variables?

```python
for i in col_names[:3]:
    print(i+' Correlation Coefficent: {}'.format(np.corrcoef(data[i],data['Profit'])[1][0]))
```

```python
for i,feature_name in enumerate(col_names[:3]):
    plt.figure(figsize=(20,20))
    plt.subplot(5,3,i+2)
    plt.scatter(data[feature_name],data['Profit'])
    plt.xlabel(feature_name)
    plt.ylabel('Profit')
    plt.show() 
```

As we can see from the graphs and the correlation coefficents R&D Spend is very positively correlated to Profit and marketing spend is decently postively correlated to profit as well.
