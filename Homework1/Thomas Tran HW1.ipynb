```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Startups.csv')
```

```python
df
```

```python
columns = list(df.columns)
columns
```

```python
categorical = ['State']
quantitative = list(set(columns).difference(categorical))
independent = list(set(columns).difference(['Profit']))
```

```python
ind_quant = list(set(quantitative).difference(['Profit']))
```

```python
quantitative+categorical
```

### Question 3

```python
for ind in independent:
    plt.hist(df[ind])
    plt.title(ind)
    plt.ylabel('Count')
    plt.show()
```

```python
for q in ind_quant:
    plt.boxplot(df[q])
    plt.title(q)
    plt.show()
```

R&D Spend is slightly right skewed.

Administration is bimodal with a slight left skew.

Marketing Spend is also bimodal but with a right skew.

Finally, State is a categorical variable

### Question 4

### Question 5

```python
for q in ind_quant:
    plt.scatter(df[q],df['Profit'])
    plt.xlabel(q)
    plt.ylabel('Profit')
    plt.show()
    print('r =',df[q].corr(df['Profit']))
```

Taking a look at the scatterplots and the corellation coefficients each of the quantitative independent variables has a positive correlation with profit. However, taking a look at the R&D Spend we can see a very strong positive correlation with Profit.

### Question 6

```python
states = ['New York', 'California', 'Florida']
```

```python
for s in states:
    temp = df.loc[df['State']==s]
    plt.hist(temp['Profit'])
    plt.title(s+' Profit')
    plt.show()
    
    print(s+':\nmax:',temp['Profit'].max(),'\naverage:',temp['Profit'].mean())
```

New York has the single highest profit of 192,261.83

However, Florida has the highest average profit of 118,774.02.
