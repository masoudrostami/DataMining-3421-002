```python
from sklearn.semi_supervised import SelfTrainingClassifier
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
iris = load_iris()
```

```python
X = iris.data
X
```

```python
Y = iris.target
Y
```

```python
len(Y)
```

```python
def get_indices(arr, elem):
    indices = []
    for i in range(len(arr)):
        if arr[i] == elem:
            indices.append(i)
    return indices

elem = 2
indices = get_indices(Y, elem)
print(indices)
```

```python
###Let's picked a few labeled data that contains all the samples from all 3 different classes
labeled_idx = [3, 6, 13, 58, 64, 67, 128, 141, 149]
unlabeled_idx = [i for i in range(len(iris.target)) if i not in labeled_idx]
unlabeled_idx
```

```python
X_labeled, y_labeled = iris.data[labeled_idx], iris.target[labeled_idx]
X_unlabeled = iris.data[unlabeled_idx]
X_labeled
```

```python
#Fit the self-training classifier on the labeled data and 
#then make predictions on the unlabeled data

clf = DecisionTreeClassifier()
self_training_clf = SelfTrainingClassifier(clf)
self_training_clf.fit(X_labeled, y_labeled)
y_unlabeled_pred = self_training_clf.predict(X_unlabeled)
y_unlabeled_pred
```

```python
X_unlabeled
```

```python
# Add the most confident predictions to the labeled data
# Make sure we don't include any weird predictions

X_new_labeled, y_new_labeled = X_unlabeled[y_unlabeled_pred != -1], y_unlabeled_pred[y_unlabeled_pred != -1]
X_labeled_new = np.concatenate((X_labeled, X_new_labeled), axis=0)
y_labeled_new = np.concatenate((y_labeled, y_new_labeled), axis=0)
```

```python
# Refit the self-training classifier on the expanded labeled data and evaluate its performance
self_training_clf.fit(X_labeled_new, y_labeled_new)
y_pred = self_training_clf.predict(iris.data)
print("Accuracy score: ", accuracy_score(iris.target, y_pred))
```
