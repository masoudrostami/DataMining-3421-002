```python
import numpy as np
import pandas as pd
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv("./heart_data.csv")
```

```python
features = df.columns

X = df[features]
y = df['cardio']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)
```

```python
train, test = train_test_split(df, test_size=0.3, random_state=0)
```

```python


features = train.columns

X = train[features]
y = train['cardio']

```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```

```python
## have cardio
## a rank of 0.5 or lower will follow the True arrow (to the left), 
## and the rest will follow the False arrow (to the right).

##  gini: 0.0 would mean all of the samples got the same result, 
## and 0.5 would mean that the split is done exactly in the middle.


## 24482 will not get cardio, 24518 will get cardio
```

```python
y_pred = dtree.predict(X_test)
```

```python
print("Train data accuracy:",metrics.accuracy_score(y_true = y_train, y_pred = dtree.predict(X_train)))
print("Test data accuracy:",metrics.accuracy_score(y_true = y_test, y_pred = y_pred))
```
