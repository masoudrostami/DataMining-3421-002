# Data Pre-Processing

```python
pd.get_dummies(df['SEX']) ## Useful one-hot function
```

```python
lst = ['SITE_ID','SUB_ID','DX_GROUP','AGE_AT_SCAN','SEX','HANDEDNESS_CATEGORY','EYE_STATUS_AT_SCAN',
       'FIQ', 'VIQ','PIQ','FIQ_TEST_TYPE','VIQ_TEST_TYPE','PIQ_TEST_TYPE',
        'ADOS_G_TOTAL','ADOS_2_TOTAL','SRS_TOTAL_T','SCQ_TOTAL','VINELAND_SUM_SCORES','RBSR_6SUBSCALE_TOTAL','RBSR_5SUBSCALE_TOTAL',
       'MASC_TOTAL_T','CBCL_1.5-5_TOTAL_T','BDI_TOTAL','WIAT-II-A_TOTAL_COMPOSITE_S','CPRS_DSM_IV_TOTAL',
       'ADI_R_SOCIAL_TOTAL_A','ADI_R_VERBAL_TOTAL_BV','ADI_R_NONVERBAL_TOTAL_BV', 'ADI_R_RRB_TOTAL_C','ADI_R_ONSET_TOTAL_D'] ## Making list of columns

DFN = df[lst]
display(DFN)
print('\nVariables:',len(DFN))
```

```python
DFN.index = DFN['SUB_ID'] #Changing index
DFN = DFN.drop(['SUB_ID','SITE_ID'], axis = 1)[DFN.EYE_STATUS_AT_SCAN.notna()] #dropping IDs as well as removing one individual with inconclusive eye scan
DFN['HANDEDNESS_CATEGORY'] = DFN['HANDEDNESS_CATEGORY'].replace(np.nan, 0) ## replacing NaNs in handedness category

numcat = ['FIQ','VIQ','PIQ','ADOS_G_TOTAL','ADOS_2_TOTAL','SRS_TOTAL_T','SCQ_TOTAL','VINELAND_SUM_SCORES','RBSR_6SUBSCALE_TOTAL','RBSR_5SUBSCALE_TOTAL',
       'MASC_TOTAL_T','CBCL_1.5-5_TOTAL_T','BDI_TOTAL','WIAT-II-A_TOTAL_COMPOSITE_S','CPRS_DSM_IV_TOTAL',
       'ADI_R_SOCIAL_TOTAL_A','ADI_R_VERBAL_TOTAL_BV','ADI_R_NONVERBAL_TOTAL_BV', 'ADI_R_RRB_TOTAL_C','ADI_R_ONSET_TOTAL_D'] ## list of columns we don't want NaNs in

## normalize columns in numcat BEFORE REPLACEMENT

DF_asd = DFN.copy()
DFN[numcat] = DFN[numcat].replace(np.nan, -1) ## Since these values are whole numbers always above 0, making the NaNs negative will theoretically let any classifier treat them differently.

DFN = pd.concat([DFN,pd.get_dummies(DFN[['FIQ_TEST_TYPE','VIQ_TEST_TYPE','PIQ_TEST_TYPE']])], axis = 1) ## using one-hot encoding for the types of IQ tests
DFN = DFN.drop(['FIQ_TEST_TYPE','VIQ_TEST_TYPE','PIQ_TEST_TYPE'], axis = 1) ## getting rid of the actual string IQ test types
```

```python
plt.matshow(DF_asd.corr())
```

```python
with pd.option_context('display.max_columns', None):
  display(DFN)

#We display all columns of our new dataset that is ready to plug into our algorithms.
```

```python
y = 2 - np.array(DFN.DX_GROUP)
y

# This is used so that the choice changes from {1,2} to {0,1}, where 0 indicates a negative diagnosis and 1 indicates a positive diagnosis.
```

```python
X = DFN.drop(['DX_GROUP'], axis = 1).to_numpy()
X
```

```python
import sklearn
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 695)
```

```python
X_train
```

# DNN

In this case, the kind of deep linear neural network I'm using is a multilayer perceptron (MLP) classifier.

```python
#import warnings
#warnings.filterwarnings('ignore') 
```

```python
from sklearn.neural_network import MLPClassifier 
from sklearn.metrics import accuracy_score, precision_score, recall_score

tup = (75,100,50,25,15,5)
clf = MLPClassifier(solver = 'sgd', hidden_layer_sizes = tup, activation = 'relu', validation_fraction = 0.1)

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy1 = accuracy_score(y_test, y_pred)
precision1 = precision_score(y_test, y_pred)
recall1 = recall_score(y_test, y_pred)
```

```python
accuracy1, precision1, recall1
```

```python
plt.plot(clf.loss_curve_)
plt.title("Loss Curve for DNN")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.show()
```

# K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

n= 50
a = []
for i in range(1,n+1):

  clf_K = KNeighborsClassifier(n_neighbors = i, weights = 'distance')

  clf_K.fit(X_train, y_train)
  y_pred = clf_K.predict(X_test)
  accuracy2 = accuracy_score(y_test, y_pred)
  precision2 = precision_score(y_test, y_pred)
  recall2 = recall_score(y_test, y_pred)
  a.append([accuracy2, precision2, recall2])
```

```python
a3 = np.array(a).transpose()
a3
```

```python
accuracy2 = a3[0,2]
precision2 = a3[1,2]
recall2 = a3[2,2]
```

```python
plt.plot(a3[0], label = "accuracy")
plt.plot(a3[1], label = "precision")
plt.plot(a3[2], label = "recall")
plt.title("Metrics")
plt.xlabel("k")
plt.ylabel("Percentage")
plt.legend()
plt.show()
```

# Decision Tree

```python
from sklearn import tree

clf_DT = tree.DecisionTreeClassifier()

clf_DT.fit(X_train, y_train)
y_pred = clf_DT.predict(X_test)
accuracy3 = accuracy_score(y_test, y_pred)
precision3 = precision_score(y_test, y_pred)
recall3 = recall_score(y_test, y_pred)
```

```python
accuracy3, precision3, recall3
```
